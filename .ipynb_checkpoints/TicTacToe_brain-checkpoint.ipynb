# Game project : Tic-Tac-Toe
### First build the game, then give it a brain!

```python
#Intro and welcome
print('Welcome to TIC-TAC-TOE\n')
l= [[' ',' ',' '],[' ',' ',' '],[' ',' ',' ']]           # defining grid
x = input('To play the game amongst yourselves press 1, to play against the machine press 2.')
print('\n')


#two player version
if int(x) == 1 :
    l= [[' ',' ',' '],[' ',' ',' '],[' ',' ',' ']]           # defining grid
    #defining arrays to check for a win as lists
    L1=l[0]
    L2 = l[1]
    L3 = l[2]
    C1 = [l[0][0],l[1][0],l[2][0]]
    C2 = [l[0][1],l[1][1],l[2][1]]
    C3 = [l[0][2],l[1][2],l[2][2]]
    D1 = [l[0][0],l[1][1],l[2][2]]
    D2 = [l[0][2],l[1][1],l[2][0]]
    Risk = [L1 , L2 , L3 , C1 , C2 , C3 , D1 , D2]            #this list of the previous one will simplify overall code 
    #turn will be the variable we use to track which player's turn it is
    turn = 1
    #end will start as False and switch to True if either player wins
    end = False
    #plays count the number of chips on the board to be able to tell when the game is a draw
    plays = 0
    
    
    # tour() is a function that enables a player to choose where he wants to place is next chip
    def tour() :
        ligne = int(input('on what line do you want to place your chip?'))-1
        col = int(input('on what col do you want to place your chip?'))-1
        turnok = 0
        player = turn
        while turnok == 0:
            if l[ligne][col] == ' ':   #si il n'y a pas encore de jeton
                if turn == 1 :         # quel jetons va être posé?
                    l[ligne][col] = 'X'
                    turnok += 1
                elif turn == 2:
                    l[ligne][col] = 'O'
                    turnok += 1
                result = 'ok'
            else:   
                print('there already is a chip in that square!')
                tour()
                result = 'no'
                break
        return result
    
    
    #winner_1() is a function that checks wether a 3 long list is a winner and returns the chip of the winner
    def winner_1(li):
        r=''
        if li[0] == li[1] and li[1] == li[2]:
            if li[0]==' ':
                r = False
            else:
                r = True
        return [r,li[0]]
    
    
    
    
    #winner() is a function that applies the function winner_1() to the lists in the Risk list
    def winner() :
        for l in Risk :
            if winner_1(l)[0] == True:
                result = winner_1(l)[1]
                break
            else :
                result = False
        return result
    
    
    
    #now the game will execute calling on the different lists, variable and functions we have defined
    for i in l:
        print(i)
    while end == False and plays <= 9:
        tour()
        for i in l:
            print(i)
        winner()
        end = winner()
        if str(end) == 'X' or str(end) == 'O' or plays == 8:
            break 
        turn=2
        plays += 1
        tour()
        for i in l:
            print(i)
        winner()
        end = winner()
        if str(end) == 'X' or str(end) == 'O' or plays == 9:
            break 
        turn=1
        plays += 1
    if end == 'X':
        print('Player 1 wins')
    elif end == 'O':
        print('Player 2 wins')
    if plays == 8:
        print('match is a draw')
    
#two player version
elif int(x) == 2 :
    import random
    l= [[' ',' ',' '],[' ',' ',' '],[' ',' ',' ']] #same as for 2 players version
    L1=l[0]                                        #same as for 2 players version
    L2 = l[1]                                      #same as for 2 players version
    L3 = l[2]                                      #same as for 2 players version
    C1 = [l[0][0],l[1][0],l[2][0]]                 #same as for 2 players version
    C2 = [l[0][1],l[1][1],l[2][1]]                 #same as for 2 players version
    C3 = [l[0][2],l[1][2],l[2][2]]                 #same as for 2 players version
    D1 = [l[0][0],l[1][1],l[2][2]]                 #same as for 2 players version
    D2 = [l[0][2],l[1][1],l[2][0]]                 #same as for 2 players version
    risk = [L1 , L2 , L3 , C1 , C2 , C3 , D1 , D2] #same as for 2 players version
    turn = 1                                       #same as for 2 players version
    end = False                                    #same as for 2 players version
    plays = 0                                      #same as for 2 players version
    #the squares list will be used by the computer to track empty squares
    squares = [[1,1],[1,2],[1,3],[2,1],[2,2],[2,3],[3,1],[3,2],[3,3]]

    # the brain function weights the risk of all lists in the risk list. If one represents a winning move for
    #the opponent, the computer will return a list holding the index of the empty square in that list as well as the 
    #index of said list in the risk list.
    def brain() :
        r = [l for l in risk if l.count('X') == 2 and l.count(' ') == 1]
        if len(r) >= 1:
            result = [r[0].index(' '),risk.index(list(r[0]))]
        else:
            result = [False]
        return result


    #the Ai function takes the outpout from the brain function and converts it to a play on the TTT grid
    def AIturn():
        lai = 0
        cai = 0
        if brain()[0] is False:
            lai = squares[random.randrange(0,len(squares))][0]
            cai = squares[random.randrange(0,len(squares))][1]
        elif brain()[1] == 0 :
            lai = 1
            cai = brain()[0]+1
        elif brain()[1] == 1 :
            lai = 2
            cai = brain()[0]+1
        elif brain()[1] == 2 :
            lai = 3
            cai = brain()[0]+1
        elif brain()[1] == 3 :
            lai = brain()[0]+1
            cai = 1
        elif brain()[1] == 4 :
            lai = brain()[0]+1
            cai = 2
        elif brain()[1] == 5 :
            lai = brain()[0]+1
            cai = 3
        elif brain()[1] == 6 :
            if brain()[0] == 0 :
                lai = 1
                cai = 1
            elif brain()[0] == 1 :
                lai = 2
                cai = 2
            elif brain()[0] == 2 :
                lai = 3
                cai = 3
        elif brain()[1] == 7 :
            if brain()[0] == 0 :
                lai = 1
                cai = 3
            elif brain()[0] == 1 :
                lai = 2
                cai = 2
            elif brain()[0] == 2 :
                lai = 3
                cai = 1
            
        return[lai,cai]
    #the tour function was modified a little compared to the 2 players version to take into account the AI's turn
    def tour() :
        turnok = 0
        player = turn
        ai = AIturn()
        while turnok == 0:
            if turn == 1 :         # quel jetons va être posé?
                ligne = int(input('on what line do you want to place your chip?'))-1
                col = int(input('on what col do you want to place your chip?'))-1
                if l[ligne][col] == ' ':   #si il n'y a pas encore de jeton
                    l[ligne][col] = 'X'
                    turnok += 1
                    squares.pop(squares.index([ligne+1,col+1]))
                else:
                    print('there already is a chip in that square!')
                    tour()
                    break
                        
            elif turn == 2:
                print(ai)
                l[ai[0]-1][ai[1]-1] = 'O'
                turnok += 1  
                squares.pop(squares.index([ai[0],ai[1]]))
            
        return 
    def winner_1(li):
        r=''
        if li[0] == li[1] and li[1] == li[2]:
            if li[0]==' ':
                r = False
            else:
                r = True
        return [r,li[0]]
    
    def winner() :
        for l in Risk :
            if winner_1(l)[0] == True:
                result = winner_1(l)[1]
                break
            else :
                result = False
        return result
    #same as 2 player version
    for i in l:
        print(i)
    while end == False and plays <= 9:
        tour()
        for i in l:
            print(i)
        winner()
        end = winner()
        if str(end) == 'X' or str(end) == 'O' or plays == 8:
            break 
        turn=2
        plays += 1
        tour()
        for i in l:
            print(i)
        winner()
        end = winner()
        if str(end) == 'X' or str(end) == 'O' or plays == 9:
            break 
        turn=1
        plays += 1
    if end == 'X':
        print('Player 1 wins')
    elif end == 'O':
        print('Player 2 wins')
    if plays == 8:
        print('match is a draw')
else: 
    print('please enter 1 or 2')
    
```
