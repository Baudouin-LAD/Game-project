```python
lnumbers = [1,2,3,4,5,6,7,8,9]
l = ['-','-','-','-','-','-','-','-','-']
L1 = [l[0],l[1],l[2]]
L2 = [l[3],l[4],l[5]]
L3 = [l[6],l[7],l[8]]
C1 = [l[0],l[3],l[6]]
C2 = [l[1],l[4],l[7]]
C3 = [l[2],l[5],l[8]]
D1 = [l[0],l[4],l[8]]
D2 = [l[2],l[4],l[6]]
R = [L1,L2,L3,C1,C2,C3,D1,D2]
C = [l[0],l[2],l[6],l[8]]
S = [l[1],l[3],l[5],l[7]]
squares = [l[0],l[1],l[2],l[3],l[4],l[5],l[6],l[7],l[8]]
turn = 0
def update():
    global L1, L2, L3, C1, C2, C3, D1, D2, R, C, S, squares
    L1 = [l[0],l[1],l[2]]
    L2 = [l[3],l[4],l[5]]
    L3 = [l[6],l[7],l[8]]
    C1 = [l[0],l[3],l[6]]
    C2 = [l[1],l[4],l[7]]
    C3 = [l[2],l[5],l[8]]
    D1 = [l[0],l[4],l[8]]
    D2 = [l[2],l[4],l[6]]
    R = [L1,L2,L3,C1,C2,C3,D1,D2]
    C = [l[0],l[2],l[6],l[8]]
    S = [l[1],l[3],l[5],l[7]]
squares = [l[0],l[1],l[2],l[3],l[4],l[5],l[6],l[7],l[8]]
def grid(l):
    print(' ',l[0],' | ',l[1],' | ',l[2],' \n ',l[3],' | ',l[4],' | ',l[5],' \n ',l[6],' | ',l[7],' | ',l[8])
    return

def check_win(x):
    #count winning move for 'x':
    update()
    cx = [l for l in R if l.count(x) == 2 and l.count('-') == 1]
    if len(cx) != 0 :
        r = [cx[0].index('-'),R.index(cx[0])]
        if r[1] == 0:
            result = cx[0].index('-')
        elif r[1] == 1:
            result = cx[0].index('-')+3
        elif r[1] == 2:
            result = cx[0].index('-')+6
        elif r[1] == 3:
            result = cx[0].index('-')*3
        elif r[1] == 4:
            result = cx[0].index('-')*3+1
        elif r[1] == 5:
            result = cx[0].index('-')*3+2 
        elif r[1] == 6:
            if cx[0].index('-') == 0:
                result = 0
            elif cx[0].index('-') == 1 :
                result = 4
            elif cx[0].index('-') == 2 :
                result = 8
        elif r[1] == 6 :
            if cx[0].index('-') == 0:
                result = 2
            elif cx[0].index('-') == 1 :
                result = 4
            elif cx[0].index('-') == 2 :
                result = 6
        
    else:
        result = False
    return result


def fork(x):
    update()
    for i in range(len(l)):
        if l[i] == '-':
            l[i] = x
            if check_win(x) != False:
                l[i] = '-'
                return i
            else:
                l[i] = '-'
    return False


def AI():
    a=0
    update()
    while a == 0 :#first play any winning move
        if turn==1:
            if l.index('X') in [0,2,6,8]:
                if l.index('X') == 0:
                    result = 8
                    a += 1
                elif l.index('X') == 2:
                    result = 6
                    a +=1
                elif l.index('X') == 6:
                    result = 2
                    a +=1
                elif l.index('X') == 8:
                    result = 0
                    a +=1
            elif l.index('X') in [1,3,5,7]:
                result = 4
                a += 1
                
        if check_win('X')!= False :
            result = check_win('X')
            a+=1
    #then block any opponent winning move:
        elif check_win('O') != False :
            result = check_win('O')
            a+=1
    #then play on a side
        elif S.count('-') > 0:
            result = S.index('-')*2+1
            a+=1
    #then check for own potential forks:
        elif fork('O') != False :
            result = fork('O')
            a+=1
    #then check for opponent's forks:
        elif fork('X') != False :
            result = fork('X')
            a+=1
    
    #then play in a corner if possible
        elif C.count('-') > 0:
            if l[0] == '-':
                result = 0
                a+=1
            elif l[2] == '-':
                result = 2
                a+=1
            elif l[6] == '-': 
                result = 6
                a+=1
            elif l[8] == '-':
                result = 8
                a+=1
    #then play in the center if possible
        else :
            result = 4
            a+=1
        l[result] = 'O'
    
    return result
    

    
        
def player():
    x= int(input('what cell do you want to play in?'))-1
    l[x] = 'X'
    return

def winner():
    cwX = [l for l in R if l.count('X')==3] 
    cwO = [l for l in R if l.count('X')==3]
    if len(cwX) != 0 :
        return 'Player wins!'
    elif len(cwO) != 0:
        return 'Computer wins!'
    else:
        return False

print('Welcome to Tic-Tac-Toe')
print('This is the grid we will be playing on:')
grid(lnumbers)
while  turn <= 9 and winner() == False:
    player()
    grid(l)
    update()
    print(l)
    turn+=1
    if turn >= 8 or winner()!=False:
        if winner()==False:
            print('The game is a draw')
            break
        else: 
            print(winner())
            break
    print('turn: ',turn)
    AI()
    turn+=1
    grid(l)
    update()
    print(l)
    winner()

```

```python
print('Welcome to Tic-Tac-Toe')
print('This is the grid we will be playing on:')
grid(lnumbers)
start = input('would you like to go first?Y/N')
if start == 'Y':
    while winner() == False:
        player()
        grid(l)
        winner()
        l[AI()]='O'
        print("computer's turn")
        grid(l)
        winner()
        
elif start == 'N':
    while winner() == False:
        print("computer's turn")
        l[AI()]='O'
        grid(l)
        winner()
        player()
        grid(l)
        winner()
print(winner())
```
